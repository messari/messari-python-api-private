# GokuStats Python API Tutorial

```python
from messari.gokustats import GokuStats
gk = GokuStats()
```

# API Structure
The Messari Python client contains a number of functions that wrap all of GokuStats's API endpoints. These include:


General
* get_datatable

Fundamentals
* get_daily_active_addresses
* get_daily_transactions
* get_daily_tvl

Stablecoins
* get_daily_stablecoin_market_cap
* get_cumulative_stablecoin_market_cap

Social
* get_daily_twitter_followers
* get_cumulative_twitter_followers

Market Data
* get_daily_prices
* get_daily_market_caps

Valuation
* get_daily_market_cap_over_active_addresses
* get_daily_market_cap_over_transactions
* get_daily_market_cap_over_tvl

Below are a few examples to showcase the functionality and types of data each function generates.
